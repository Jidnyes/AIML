```python
#importing all necessary files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
#accessing the DataSet
file_path = "Titanic-Dataset.csv"

# Load the latest version
hf_dataset = kagglehub.load_dataset(
  KaggleDatasetAdapter.HUGGING_FACE,
  "yasserh/titanic-dataset",
  file_path,
  # Provide any additional arguments like
  # sql_query, hf_kwargs, or pandas_kwargs. See
  # the documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterhugging_face
)

print("Hugging Face Dataset:", hf_dataset)
#df = pd.read_csv('titanic.csv') alternative of above file
```

```python
# Displaying the values and information about null value
# Convert Hugging Face dataset to pandas DataFrame for easier manipulation
df = pd.DataFrame(hf_dataset)

print("basic information of data")
print(df.info())

print("show 1st 5 rows ")
print(df.head())

print("show last 5 rows ")
print(df.tail())

print("show row and columns (row,columns)")
print(df.shape)

print('checks for missiong value')
print(df.isnull().sum())

print("total null values :")
print(df.isnull().sum().sum())
```

```python
# filling null values
# fill null values with specified values ex., 0 here
df1=df.fillna(value=0) # all empty values are replaced by 0 or any number or value assigned to value variable
df1 # 0 null values now

# filling null values with the above row value in the dataset ex., if age of id 5 is 18 and age of id 6 is null then it will be assigned as of 5 i.e., 18
df2=df.fillna(method="pad")
df2 # may null value contain if the above column is also empty

# filling null values with the above row value in the dataset ex., if age of id 6 is 18 and age of id 5 is null then it will be assigned as of 5 i.e., 18
df3=df.fillna(method="bfill")
df3 # may null value contain if the below column is also empty

# filling null value with the next column value
df4 = df.fillna(method="bfill", axis=1)
df4

# filling null value with the below column value
df5 = df.fillna(method="pad", axis=1)
df5

# filling null columns columnwise with the specified value
df6 = df.fillna({"Age": 0, "Fare": 0, "Embarked" : "unknown"})
df6

# filling null value with the mean of column (only for numeric values)
df7 = df.fillna(df['Fare'].mean()) # mean can be changed into median,max,min
df7

# median is the middle value , mean is avg
```

```python
# drop the null values ( whole row )
df8=df.dropna()
df8
# drop the row if a single value is also null
df9=df.dropna(how="any")
df9

# drop the row if all values are null
df10=df.dropna(how="all")
df10
```

```python
# replacing a null or any other value with a specified value using replace()
import numpy as np
df11=df.replace(to_replace=np.nan,value=13) # replacing null value into 13
df11

#replacing other value
df12=df.replace(to_replace="others",value="unknown")
df12
```

```python
#interpolate() which gives the mean of 2 values i.e., if id 5 age =10 and if id 6 age=nan and id 7 age= 15 then id 6 age =13 ( mean of above and below values )
df['Fare']=df['Fare'].interpolate(method='linear')
df
```
